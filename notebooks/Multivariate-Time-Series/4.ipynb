```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8-darkgrid")
```

```python
preds_df = pd.read_csv("../../models/lstm/preds.csv")

# Choose best run
run_id = "ws20_scalerRobustScaler_lossmse_bs32_ep200"
best_preds = preds_df[preds_df["run_id"] == run_id].copy()
```

```python
plt.figure(figsize=(10, 5), dpi=120)
plt.plot(best_preds["actual"].values, label="Actual Price", color="blue")
plt.plot(best_preds["predicted"].values, label="Predicted Price", color="red")
plt.title(f"Predictions from {run_id}")
plt.xlabel("Time Step (Days)")
plt.ylabel("Price (USD)")
plt.legend()
plt.tight_layout()
plt.show()
```
